```python
#create a Dataset called Customers
Customers = {
  "customer_id": ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10'],
  "customer_name": ['JOYCE EZEONWU', 'JAYDEN PETERS', 'JANELLE PINK', 'ZANE AFAMA', 'PEARL NDUKWE', 'AKUNNA NDUKWE', 'IVAN REGINALD', 'INDIRA OLANIPEKUN', 'IFUNANYA ONUORAH','CHRISTOPHER ONYEKWELU'],
  "customer_address": ['SURULERE', 'YABA', 'LEKKI', 'IKOYI', 'IKOTUN', 'AJAH', 'GBAGADA', 'IKEJA', 'IKOTA', 'VICTORIA ISLAND'],
  "customer_phone_number": ['08133667524 ', '09035617890 ', '08035473561', '09055098871', '08160126024', '08076136249', '07053108170', '09057544364', '08143389725', '08097992016']
            }
df_customers = pd.DataFrame(Customers)
df_customers
```

```python
#Carrying out an inner merge operation between the orders table and products table
df_ordersproducts = df_order.merge(df_products, how='inner', indicator = True)
df_ordersproducts.head()
```

What I observed: The inner merge operation brought the two datasets together into one, and aligned the rows from each based on common attributes or columns. When the default value of the 'how' parameter is set to 'inner', a new DataFrame is generated from the intersection of the left and right DataFrames. Therefore, if a 'order_id' is missing in one of the tables, it would not be in the merged DataFrame. All the other ids are not part of the merged DataFrames since they do not intersect on both tables.

```python
#Performing another inner merge operation with the customers table using the resulting table
df_all = df_ordersproducts.merge(df_customers, how='inner')
df_all.head()
```

Observation: It returned a dataframe with only those rows that have common characteristics.

```python
#Performing an inner merge operation between customers and orders tables. , on customer_id
df_customersorders = df_customers.merge(df_order, how='inner', left_on = 'customer_id', right_on = 'customer_id', suffixes = ("_customers", "_order"))
df_customersorders.head()
```

```python
#Performing an outer merge between customers and products tables. 

df_customersproducts = pd.merge(df_customers,df_products,how ='outer')
df_customersproducts
```

```python
#Performing an outer merge between the customers and products tables. using left_index,right_index
df_customersproducts = pd.merge(df_customers, df_products, left_index=True, right_index=True, how ='outer', indicator = True)
df_customersproducts.head()
```

Observation: The customers and products dataframes have no common colums to perform merge on because the matching indexes were set to default 'False" as seen in the error message.
However after setting the matching indexes as 'TRUE' (left_index, right_index), the customers and products dataframes were able to merge as seen above.

```python
#Performing a left merge between orders and customers tables. 

df_orderscustomers = pd.merge(df_order,df_customers, how='left', on='customer_id')
df_orderscustomers
```

Observation: Using the customer_ID, the customer and product dataframes were successfully merged. All the records from the first dataframe were displayed, irrespective of whether the keys in the first dataframe can be found in the second dataframe. Whereas, for the second dataframe, only the records with the keys in the second dataframe that can be found in the first dataframe were displayed. There are some NaN values which show missing values in 'order' dataframe which do not match the customer dataframe.

```python
#Performing a right merge between product and customers tables
df_productscustomers = pd.merge(df_products, df_customers, left_index=True, right_index=True, how = 'right')
df_productscustomers.head()
```

The product and customer dataframes were merged using matching index. All the records from the second dataframe were displayed. However, only the records with the keys in the first dataframe that can be found in the second dataframe were displayed.The data that had matching details were successfully merged.
